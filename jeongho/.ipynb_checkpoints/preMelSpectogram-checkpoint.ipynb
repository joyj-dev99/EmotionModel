```python
현재까지
TS1 
완료

sr=16000 (샘플링 속도)와 n_mels=128 (Mel 필터 개수) 선택 이유
일반적인 설정
sr
16000 Hz는 음성 처리에서 널리 사용되는 표준 샘플링 속도
인간의 음성 대역은 약 300 Hz ~ 3400 Hz 사이에 집중되어 있으며, 16000 Hz로 샘플링하면 Nyquist 이론에 의해 이 대역을 충분히 커버
16000 Hz는 오디오 데이터 크기를 줄이고 연산량을 감소시키면서도 음성 특징을 충분히 보존

n_mels
n_mels=128은 MobileNetV3의 입력 크기(예: 224x224)로 쉽게 조정 가능

입력크기(224×224) 선택 이유
ImageNet 데이터셋 표준
ImageNet 데이터셋은 사전 학습 모델의 학습에 자주 사용되며, 기본 이미지 크기가 224×224.
모바일넷v3 ImageNet 데이터셋에서 학습된 가중치를 사용하며, 이 크기를 유지해야 가중치를 활용
2의 배수로 크기가 잘 나뉘며, 풀링(pooling) 연산 및 스트라이드(stride)와 호환이 잘 됨.
최종 특징 맵 크기를 적절히 줄여주는 역할
```

```python
import os
import time
import librosa
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize

input_dir = "/data/audio1/unzipped_TS1/TS1/folder_1/folder_1/"
output_dir = "/data/PreTS1/"
os.makedirs(output_dir, exist_ok=True)

def extract_mel_spectrogram(wav_path, sr=16000, n_mels=128):
    audio, sr = librosa.load(wav_path, sr=sr)
    mel_spec = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=n_mels)
    log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)
    return log_mel_spec

def resize_spectrogram(mel_spec, target_shape=(224, 224)):
    return resize(mel_spec, target_shape, mode="constant", anti_aliasing=True)

def save_spectrogram_as_image(mel_spec, output_path):
    plt.figure(figsize=(2.24, 2.24))
    plt.axis("off")
    plt.imshow(mel_spec, aspect="auto", origin="lower", cmap="viridis")
    plt.savefig(output_path, bbox_inches="tight", pad_inches=0)
    plt.close()

existing_files = set(os.listdir(output_dir))
count =0
for root, _, files in os.walk(input_dir):
    
    for file in files:
        count = count +1
        if file.endswith(".wav"):
            if f"{os.path.splitext(file)[0]}.png" in existing_files:
                print(f"Skipping (duplicate): {os.path.splitext(file)[0]}.png")
                if count %500 == 0:
                    time.sleep(1)
                continue
                
            wav_path = os.path.join(root, file)
            mel_spec = extract_mel_spectrogram(wav_path)
            resized_mel = resize_spectrogram(mel_spec, target_shape=(224, 224))
            output_path = os.path.join(output_dir, f"{os.path.splitext(file)[0]}.png")

            
                
            save_spectrogram_as_image(resized_mel, output_path)
            print(f"Saved: {output_path}")

print('end')
```
