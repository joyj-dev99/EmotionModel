```python
import os
import time
import librosa
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize


def process_folder(folder_path):
    input_dir = folder_path
    output_dir = "/data/preTS"
    os.makedirs(output_dir, exist_ok=True)
    
    def extract_mel_spectrogram(wav_path, sr=16000, n_mels=128):
        audio, sr = librosa.load(wav_path, sr=sr)
        mel_spec = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=n_mels)
        log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)
        return log_mel_spec
    
    def resize_spectrogram(mel_spec, target_shape=(224, 224)):
        return resize(mel_spec, target_shape, mode="constant", anti_aliasing=True)
    
    def save_spectrogram_as_image(mel_spec, output_path):
        plt.figure(figsize=(2.24, 2.24))
        plt.axis("off")
        plt.imshow(mel_spec, aspect="auto", origin="lower", cmap="viridis")
        plt.savefig(output_path, bbox_inches="tight", pad_inches=0)
        plt.close()
    
    existing_files = set(os.listdir(output_dir))
    count =0
    for root, _, files in os.walk(input_dir):
        
        for file in files:
            if count %1000 == 0:
                time.sleep(1)
            count = count +1
            if file.endswith(".wav"):
                if f"{os.path.splitext(file)[0]}.png" in existing_files:
                    print(f"Skipping (duplicate): {os.path.splitext(file)[0]}.png")
                    if count %1000 == 0:
                        time.sleep(0.1)
                    continue
                    
                wav_path = os.path.join(root, file)
                mel_spec = extract_mel_spectrogram(wav_path)
                resized_mel = resize_spectrogram(mel_spec, target_shape=(224, 224))
                output_path = os.path.join(output_dir, f"{os.path.splitext(file)[0]}.png")
                
                save_spectrogram_as_image(resized_mel, output_path)
                print(f"Saved: {output_path}")
    
    print('end')
    print(f"Processing folder: {folder_path}")

def find_and_process_folders(base_path):
    for root, dirs, files in os.walk(base_path):
        wav_files = [f for f in files if f.endswith('.wav')]
        if wav_files:
            process_folder(root)

# 시작 경로 설정
base_path = "/data/audio1/unzipped_TS1/TS1/folder_1/folder_2/"
find_and_process_folders(base_path)
```
